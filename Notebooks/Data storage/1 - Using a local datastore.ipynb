```python
import sentier_data_tools as sdt
import pandas as pd
```

```python
df = pd.read_excel(
    sdt.example_data_dir / "electrolyzers.xlsx"
)
df
```

```python
datapackage = sdt.Datapackage(
    name="electrolyzer_data",
    metadata={},
    version=1
).save()
```

```python
record = sdt.Record(
    data=df,
    columns=list(df.columns),
    units=[],
    datapackage=datapackage
).save()
```
